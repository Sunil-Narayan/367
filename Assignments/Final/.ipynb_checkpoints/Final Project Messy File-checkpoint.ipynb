Now that we have all of the relevent mortality rates in isolation, it is time to merge them together. Note that the adult rates seem higher than the child rates. This is because the adult rates are total deaths across a 45 year span. To give better intuition, I may use a normalized-per-year rate.

```python
#Mortality Frame
MF = IMF.merge(CMF)
MF = MF.merge(AMF)
MF
```

The above frame shows the rate of moratlity in 2016 across all countries in three phases: Infant(age 0-1), Child(age 1-5), and Adult(age 15-60). For better intution, the following table is normalized to a per-year basis. 

```python
#Normalized Mortality Frame
NMF = MF
NMF["Child Mortality Rate"] = NMF["Child Mortality Rate"]/5
NMF["Adult Mortality Rate"] = NMF["Adult Mortality Rate"]/45
NMF
```

```python
#Resetting Mortality Frame
MF = IMF.merge(CMF)
MF = MF.merge(AMF)
MF
```

At this point, I have realized that finding the intersection of the GDP frame and the mortality frame leades to an incomplete picture. The reason for this is a difference in naming across the frames which leads to problems when merging. For example, if one frame has "United States" as a country and the other has "United States of America", merging on country will be a problem between them. I will have to fix this manually.

```python
GDPCountries = set(CF.Country)
MortalityCountries = set(MF.Country)
Difference = GDPCountries.symmetric_difference(MortalityCountries)
GDPCountries.intersection(Difference)
```

The list above shows the difference in the sets. I will now manually fix these differences. Note: Andorra, Bahamas, Monaco, Saint Kitts and Nevis and San Marino do not have full data sets (most likely due to size), and thus be dropped from this analysis

```python
MF.loc[MF["Country"] == "Czechia", "Country"] = "Czech Republic"
MF.loc[MF["Country"] == "Korea, Rep.", "Country"] = "Republic of Korea"
MF.loc[MF["Country"] == "Slovak Republic", "Country"] = "Slovakia"
MF.loc[MF["Country"] == "United Kingdom of Great Britain and Northern Ireland", "Country"] = "United Kingdom"
```

Rechecking the intersection

```python
GDPCountries = set(CF.Country)
MortalityCountries = set(MF.Country)
Difference = GDPCountries.symmetric_difference(MortalityCountries)
GDPCountries.intersection(Difference)
```

These are all of the dropped countries. We are now ready to analyze the 56 remaining countries

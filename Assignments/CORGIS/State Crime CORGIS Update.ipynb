```python
import pandas as pd
import numpy as np
```

```python
addingFrame = pd.read_csv("estimated_crimes_1979_2019.csv")
```

```python
addingFrame.head()
```

```python
stateFrame= pd.read_csv("state_crime.csv")
```

```python
stateFrame.head()
```

```python
addingFrame.rename(columns = {"state_name":"State", "population":"Data.Population"}, inplace = True )
```

```python
addingFrame.columns
```

```python
mergeFrame = addingFrame.loc[:, ['year', 'State', 'Data.Population', 'violent_crime',
       'homicide', 'rape_legacy', 'rape_revised', 'robbery',
       'aggravated_assault', 'property_crime', 'burglary', 'larceny',
       'motor_vehicle_theft']]
```

```python
stateFrame.columns
```

```python
mergeFrame = mergeFrame.rename(columns = {"year":"Year", "homicide":"Data.Totals.Violent.Murder", "rape_legacy":"Data.Totals.Violent.Rape", "rape_revised":"Data.Totals.Violent.RapeR", "Robbery":"Data.Totals.Violent.Robbery", "aggravated_assault":"Data.Totals.Violent.Assault", "property_crime":"Data.Totals.Property.All", "burglary":"Data.Totals.Property.Burglary", "larceny":"Data.Totals.Property.Larceny", "motor_vehicle_theft":"Data.Totals.Property.Motor", "robbery":"Data.Totals.Violent.Robbery", "violent_crime":"Data.Totals.Violent.All"})
```

```python
mergeFrame.State = mergeFrame.State.replace(to_replace = np.NaN, value = "United States")
```

```python
mergeFrame = mergeFrame.replace(to_replace = np.NaN, value = 0)
```

```python
mergeFrame = mergeFrame[mergeFrame.Year>=2013]
```

```python
mergeFrame["Data.Totals.Violent.Rape"] = mergeFrame["Data.Totals.Violent.Rape"]+mergeFrame["Data.Totals.Violent.RapeR"]
```

```python
mergeFrame = mergeFrame.drop(columns = "Data.Totals.Violent.RapeR")
```

```python
rateModifier = 100000/mergeFrame["Data.Population"]
```

```python
rateValues = pd.DataFrame({"State":mergeFrame.State, "Year":mergeFrame.Year, "Data.Rates.Property.All":(mergeFrame["Data.Totals.Property.All"]*rateModifier), "Data.Rates.Property.Burglary":(mergeFrame["Data.Totals.Property.Burglary"]*rateModifier), "Data.Rates.Property.Larceny":(mergeFrame["Data.Totals.Property.Larceny"]*rateModifier),"Data.Rates.Property.Motor":(mergeFrame["Data.Totals.Property.Motor"]*rateModifier),"Data.Rates.Violent.All":(mergeFrame["Data.Totals.Violent.All"]*rateModifier),"Data.Rates.Violent.Assault":(mergeFrame["Data.Totals.Violent.Assault"]*rateModifier),"Data.Rates.Violent.Murder":(mergeFrame["Data.Totals.Violent.Murder"]*rateModifier),"Data.Rates.Violent.Rape":(mergeFrame["Data.Totals.Violent.Rape"]*rateModifier),"Data.Rates.Violent.Robbery":(mergeFrame["Data.Totals.Violent.Robbery"]*rateModifier)})
```

```python
rateValues.head()
```

```python
rateValues.shape
```

```python
mergeFrame.shape
```

```python
finalMergeFrame = pd.merge(mergeFrame, rateValues)
```

```python
finalMergeFrame.tail()
```

```python
stateFrame.shape
```

```python
finalMergeFrame.shape
```

```python
completedSet = stateFrame.append(finalMergeFrame)
```

```python
completedSet.tail()
```

```python
completedSet
```

```python
completedSet = completedSet.sort_values(by=["State", "Year"])
```

```python
completedSet.to_csv("State Crime Updated through 2019")
```
